## 레드 와인의 화화적 성분과 등급을 기록한 데이터
https://archive.ics.uci.edu/dataset/186/wine+quality

### 라이브러리 불러오기

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

```python
red_df = pd.read_csv('./wine+quality/winequality-red.csv', sep = ';')
white_df = pd.read_csv('./wine+quality/winequality-white.csv', sep = ';')
```

```python
red_df.columns == white_df.columns
```

```python
red_df.shape, white_df.shape
```

```python
red_df.head(3)
```

```python
red_df.info()
```

```python
red_X = red_df.iloc[:, : -1].to_numpy()
red_y = pd.get_dummies(red_df.iloc[:, -1]).to_numpy()

# train-test 분리
X_train, X_test, y_train, y_test = \
train_test_split(red_X
                 , red_y
                 , test_size = 0.3
                 , random_state = 1
                )
```

```python
X_train.shape, y_train.shape
```

### 문제 정의 : 와인의 품질 속성을 분류하기
### 모형 설계
▶ Sequential()
- 레이어 1 : 뉴런 50, 활성화 함수(relu)
- 레이어 2 : 뉴런 50, 활성화 함수(relu)
- 레이어 3 : 뉴런 30, 활성화 함수(relu)
- 레이어 4 : 출력 6, 활성화 함수(softmax)

▶ compile()
- 최적화 : adam
- 손실함수 : 'categorical_crossentropy'
- 평가지표 : 'accuracy'

### 1. 모델 정의

```python
model = Sequential()
model.add(Input(shape = (11,))) # 입력층 11개 피쳐 지정 # built를 True로 줌
model.add(Dense(50, activation = 'relu')) # 은닉층 레이어1
model.add(Dense(50, activation = 'relu')) # 은닉층 레이어2
model.add(Dense(30, activation = 'relu')) # 은닉층 레이어3
model.add(Dense(6, activation = 'softmax')) # 출력층 레이어

model
```

### 2. 컴파일

```python
model.compile(optimizer = 'adam'
            , loss = 'categorical_crossentropy'
            , metrics = ['accuracy']
            )
model
```

### 3. 모델 훈련(학습)

```python
model.fit(X_train
          , y_train
          , epochs = 200
          , batch_size = 50
          , verbose = True
          )
```

### 4. 평가

```python
pred = np.argmax(model.predict(X_test), axis =1 )
pred
```

### 실제값과 비교

```python
# 실제 값
real_value = np.argmax(y_test, axis = 1)

pd.crosstab(real_value, pred, margins = True) # 혼동 행렬
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print('loss : ', loss)
print('accuracy : ', accuracy)
```
